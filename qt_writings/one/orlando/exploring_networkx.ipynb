```python
G = nx.Graph()
```

```python
G.add_node("woman")
```

```python
G.add_nodes_from(['reached',
 'till',
 'friend',
 'word',
 'moment',
 'saw',
 'always',
 'could',
 'cried',
 'sailor',
 'wit',
 'scarcely',
 'petticoat',
 'go',
 'servant',
 'conclusion'])
```
